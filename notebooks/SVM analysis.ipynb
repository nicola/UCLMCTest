```python
%matplotlib inline
import sys
import os
import matplotlib.pyplot as plt
sys.path.append(os.path.abspath('../../'))
import pandas as pd

from UCLMCTest.grading import grading

def yy(vector):
    result = []
    for y in vector:
        if y == 0:
            result+=[1,0,0,0]
        elif y == 1:
            result+=[0,1,0,0]
        elif y == 2:
            result+=[0,0,1,0]
        elif y == 3:
            result+=[0,0,0,1]
    return result
    

pickles = os.path.dirname(os.path.realpath(".")) + "/pickles/"

bowcoref = {}
bowcoref["train"] = pd.read_pickle(pickles + "_bowcoref_mc160.train.pickle")
bowcoref["dev"] = pd.read_pickle(pickles + "_bowcoref_mc160.dev.pickle")
bowcoref["test"] = pd.read_pickle(pickles + "_bowcoref_mc160.test.pickle")

hypbow = {}
hypbow["train"] = pd.read_pickle(pickles + "_hypbow_mc160.train.pickle")
hypbow["dev"] = pd.read_pickle(pickles + "_hypbow_mc160.dev.pickle")
hypbow["test"] = pd.read_pickle(pickles + "_hypbow_mc160.test.pickle")

bowall1 = {}
bowall1["train"] = pd.read_pickle(pickles + "_bowall1_mc160.train.pickle")
bowall1["dev"] = pd.read_pickle(pickles + "_bowall1_mc160.dev.pickle")
bowall1["test"] = pd.read_pickle(pickles + "_bowall1_mc160.test.pickle")

bowall2 = {}
bowall2["train"] = pd.read_pickle(pickles + "_bowall2_mc160.train.pickle")
bowall2["dev"] = pd.read_pickle(pickles + "_bowall2_mc160.dev.pickle")
bowall2["test"] = pd.read_pickle(pickles + "_bowall2_mc160.test.pickle")

hypselect = {}
hypselect["train"] = pd.read_pickle(pickles + "_hypselect_mc160.train.pickle")
hypselect["dev"] = pd.read_pickle(pickles + "_hypselect_mc160.dev.pickle")
hypselect["test"] = pd.read_pickle(pickles + "_hypselect_mc160.test.pickle")

hypselect2 = {}
hypselect2["train"] = pd.read_pickle(pickles + "_hypselect2_mc160.train.pickle")
hypselect2["dev"] = pd.read_pickle(pickles + "_hypselect2_mc160.dev.pickle")
hypselect2["test"] = pd.read_pickle(pickles + "_hypselect2_mc160.test.pickle")

baseline = {}
baseline["train"] = pd.read_pickle(pickles + "Baseline (BOW)_mc160.train.pickle")
baseline["dev"] = pd.read_pickle(pickles + "Baseline (BOW)_mc160.dev.pickle")
baseline["test"] = pd.read_pickle(pickles + "Baseline (BOW)_mc160.test.pickle")

filter160 = {}
filter160["train"] = pd.read_pickle(pickles + "_filter160_mc160.train.pickle")
filter160["dev"] = pd.read_pickle(pickles + "_filter160_mc160.dev.pickle")
filter160["test"] = pd.read_pickle(pickles + "_filter160_mc160.test.pickle")

yy_train = pd.read_pickle(pickles + "ymc160.train.pickle")
yy_dev = pd.read_pickle(pickles + "ymc160.dev.pickle")
yy_test = pd.read_pickle(pickles + "ymc160.test.pickle")

y_train = yy(yy_train)
y_dev = yy(yy_dev)
y_test = yy(yy_test)
```

```python
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.svm import SVC
import numpy as np
from itertools import chain, combinations

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))


tuned_parameters = [{'kernel': ['linear'], 'C': [1.0*x/10 for x in range(20,50)]}]
```

```python
features_p = list(powerset([hypbow, bowcoref, baseline, bowall1, bowall2, hypselect, hypselect2]))[1:]
features_s = list(powerset(["hypbow", "bowcoref", "baseline", "bowall1", "bowall2", "hypselect", "hypselect2"]))[1:]
features_s
```

```python
X_tests = [
    np.array(zip(*[v["test"] for v in features]))
    for features in features_p
]

X_trains = [
    np.array(zip(*[v["train"] for v in features]))
    for features in features_p
]

X_devs = [
    np.array(zip(*[v["dev"] for v in features]))
    for features in features_p
]
```

```python
clfs = []
for X_train in X_trains:
    clf = GridSearchCV(SVC(probability=True), tuned_parameters)
    clf.fit(X_train, y_train)
    clfs.append(clf)
```

```python
SVMs = []
for x, _ in enumerate(features_s):
    SVMs.append([features_s[x], X_devs[x], X_tests[x], clfs[x]])
```

```python
y_pred_devs = []
y_pred_tests = []
best_params = []

for name, X_dev, X_test, clf in SVMs:
    y_pred_devs.append([x[1] for x in clf.predict_proba(X_dev)])
    y_pred_tests.append([x[1] for x in clf.predict_proba(X_test)])
    best_params.append(clf.best_params_)
    
Ys = []
for x, _ in enumerate(features_s):
    Ys.append([features_s[x], y_pred_devs[x], y_pred_tests[x], best_params[x]])
```

```python
grades = []
for name, y_pred_dev, y_pred_test, best_param in Ys:
    grades_dev = grading(y_pred_dev, yy_dev)
    grades_test = grading(y_pred_test, yy_test)
    grades.append(dict(dev=sum(grades_dev)/len(grades_dev), test=sum(grades_test) / len(grades_test), name=name))
    
pd.DataFrame(grades).sort(["dev"], ascending=[0])
```
